```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.sample(10000)
y = np.random.sample(10000)

fig, ax = plt.subplots()
ax.hist(x, bins=20)
plt.show()
```

## Generating Normal distribution:

$\xi, \eta \sim R(0, 1)$ then

$X = (-2 \ln \xi)^{1/2} \cos{2\pi \eta}$

$Y = (-2 \ln \eta)^{1/2} \cos{2\pi \xi}$

```python
plt.figure(figsize(13,8))

n_bins = 20
x_normal = np.sqrt(-2 * np.log(x)) * np.cos(2*np.pi*y)
y_normal = np.sqrt(-2 * np.log(y)) * np.cos(2*np.pi*x)
x_test = np.sqrt(-2 * np.log(x)) * np.cos(2*np.pi*x)
y_test = np.sqrt(-2 * np.log(y)) * np.cos(2*np.pi*y)

fig, axs = plt.subplots(2, 2, sharey=True)

axs[0,0].hist(x_normal, bins=n_bins)
axs[0,1].hist(y_normal, bins=n_bins)
axs[1,0].hist(x_test, bins=n_bins)
axs[1,1].hist(y_test, bins=n_bins)
plt.show()
```

## Central Limit Theorem

```python
plt.figure(figsize(6,4))
n = 10
n_samples = 10000
X = np.random.uniform(-1, 1, (n_samples, n))
sample = []
for i in range(n_samples):
    sample.append(np.sum(X[i])/np.sqrt(n))
    
fig, ax = plt.subplots()
ax.hist(sample, bins=20)
plt.show()
```
